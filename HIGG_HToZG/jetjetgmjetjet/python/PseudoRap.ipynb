```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def ETA(theta):
    return -np.log(np.tan(theta/2))
```

```python
def dR(theta,theta1,phi,phi1):
    eta = ETA(theta)
    eta1 = ETA(theta1)
    return(np.sqrt((eta-eta1)**2 + (phi-phi1)**2))
```

```python
theta = np.arange(0.01,np.pi, 0.01)
eta = -np.log(np.tan(theta/2))
Deta = [[0 for _ in eta] for _ in eta]
for i in range(len(theta)):
    for j in range(len(theta)):
        Deta[i][j] = -np.log(np.tan(theta[i]/2))+np.log(np.tan(theta[j]/2))
        if np.abs(Deta[i][j]) >= 2.4:
            Deta[i][j] = 0
plt.figure()
plt.plot(theta*(180/np.pi),eta)
plt.plot(theta*(180/np.pi),eta*0+2.4)
plt.plot(theta*(180/np.pi),eta*0-2.4)
ax = plt.gca()
ax.set_xlim([10,170])
plt.grid(linestyle='--')


#X,Y = np.meshgrid(eta,eta)
X,Y = np.meshgrid(theta*(180/np.pi),theta*(180/np.pi))
plt.figure(figsize = (10,10))

CS = plt.contourf(X,Y,Deta,30,cmap='RdGy')
CS2 = plt.contour(CS, levels=[2.4,1,-1,-2.4],#CS.levels[::2],
                  colors='b',                  
                 )
cbar = plt.colorbar(CS)
cbar.add_lines(CS2)
plt.show()
```

```python
def delta_phi(obj1,obj2):
    #d_phi =  obj1 - obj2
    #if   d_phi >= 2*np.pi: d_phi -= 2*np.pi
    #elif d_phi <        0: d_phi += 2*np.pi
    #if d_phi > np.pi: 
    #    d_phi = 2*np.pi - d_phi
    #elif d_phi < -np.pi: 
    #    d_phi = 2*np.pi + d_phi
    d_phi =  np.abs(obj1 - obj2)
    if d_phi > np.pi:
        d_phi = 2*np.pi - d_phi
    return d_phi
```

```python
phi = np.arange(0,2*np.pi,step = 0.1)
X,Y = np.meshgrid(phi,phi)
Dphi = [[0 for _ in phi] for _ in phi]
for i in range(len(phi)):
    for j in range(len(phi)):
        Dphi[i][j] = delta_phi(phi[i],phi[j])
        
plt.contourf(X,Y,Dphi,60)#,cmap = "RdGy")
plt.colorbar()
```
